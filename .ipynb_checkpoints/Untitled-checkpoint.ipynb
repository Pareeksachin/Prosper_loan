```python
import pandas as pd

# Assuming you have your dataset loaded into a DataFrame called 'prosper'
# If you don't have a target variable, you can skip the next line
target_variable = 'LoanStatus'  # Replace with the name of your target variable

# Calculate the correlation matrix
correlation_matrix = prosper.corr()

# If you want to visualize the correlation matrix:
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

```

```python
# Identify features with high correlation to the target variable
target_correlation_threshold = 0.1  # Adjust as needed
highly_correlated_features = correlation_matrix[abs(correlation_matrix[target_variable]) > target_correlation_threshold][target_variable]

# Print the features with high correlation to the target variable
print("Features with High Correlation to", target_variable)
print(highly_correlated_features)

# Identify features with high pairwise correlation (excluding the target variable)
pairwise_correlation_threshold = 0.7  # Adjust as needed
highly_correlated_pairs = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > pairwise_correlation_threshold:
            highly_correlated_pairs.add((correlation_matrix.columns[i], correlation_matrix.columns[j]))

# Print the highly correlated pairs of features
print("Highly Correlated Feature Pairs (excluding the target variable):")
print(highly_correlated_pairs)
```
