Let’s examine the credit rating of loans when they went live. To measure the credit rating we can use the Credit Grade variable. This variable is populated only for listings before 2009 (before the SEC cease). The Credit Grade is expressed in an alphabetical grade scale where AA is the best possible credit rating for a loan taker and HR is the worst possible rating (NC means No Credit).

```python
# Define the order of credit grades
credit_grade_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR', 'NC']

# Create a bar plot for credit grades
plt.figure(figsize=(8, 6))
sns.countplot(data=PLD[~PLD['CreditGrade'].isna()], x='CreditGrade', order=credit_grade_order)
plt.xlabel("Credit Grade")
plt.ylabel("Count")
plt.title("Credit Grade Distribution")
plt.show()

# Display the count of each credit grade
pd.crosstab(index=PLD['CreditGrade'], columns='Count')
```

Most loan takers before 2009 had a credit rating of C and D. Note that we removed NA's from the graph because essentially NA's in the Credit Grade variable correspond to loan takers after July 2009.

As for the loan takers after July 2009, we are going to use the ProsperRating (Alpha) variable. The ProsperRating is a proprietary system developed by PROSPER after the SEC cease that "allows potential investors to easily consider a loan application’s level of risk because the rating represents an estimated average annualized loss rate range to the investor." So it seems reasonable to use it for measuring the credit risk of each loan taker as it fulfills the same role as the Credit Grade variable.

```python
# Define the order of Prosper Ratings
prosper_rating_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']

# Create a bar plot for Prosper Ratings
plt.figure(figsize=(8, 6))
sns.countplot(data=PLD[~PLD['ProsperRating (Alpha)'].isna()], x='ProsperRating (Alpha)', order=prosper_rating_order)
plt.xlabel("Prosper Rating")
plt.ylabel("Count")
plt.title("Prosper Rating Distribution")
plt.show()

# Display the count of each Prosper Rating
pd.crosstab(index=PLD['ProsperRating (Alpha)'], columns='Count')
```

The Prosper Rating follows a similar grading scale as the Credit Rating.

Most loan takers after 2009 had a credit rating of B and C, and few loan takers had a credit rating of AA.

Bivariate Plots Section

We will name the new variable as Defaulted and it will take only two values: either “Yes”" or “No”. We will include to the “Yes” category the Defaulted and Chargedoff loans from the LoanStatus variable and to the “No” category the Completed loans of the same variable. We will omit the other categories from the LoanStatus variable because we don’t know their outcome yet (if the loan will default or not).

As a consequence we will drop the data for all loans that don’t belong to the aforementioned categories. The total number of loans that we are going to work with is 55.084. Let’s have a first look at the percentage of defaulted loans in the newly created dataset:

```python
# Bar plot for LoanStatus distribution
plt.figure(figsize=(10, 6))
loan_status_counts = PLD['LoanStatus'].value_counts()
plt.bar(loan_status_counts.index, loan_status_counts.values, color='seagreen')
plt.xlabel("LoanStatus")
plt.ylabel("Count")
plt.title("Loan Status Distribution")

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

plt.show()
```

```python

# Count the number of rows with specific LoanStatus values
completed_defaulted_chargedoff_count = len(PLD[(PLD['LoanStatus'] == "Completed") | (PLD['LoanStatus'] == "Defaulted") | (PLD['LoanStatus'] == "Chargedoff")])
print("Number of Rows with LoanStatus 'Completed', 'Defaulted', or 'Chargedoff':", completed_defaulted_chargedoff_count)
```

```python
# Filter rows where LoanStatus is "Chargedoff", "Defaulted", or "Completed"
PLD_def_or_no = PLD[PLD['LoanStatus'].isin(["Chargedoff", "Defaulted", "Completed"])].copy()

# Create a new column "Defaulted" as a factor based on LoanStatus
PLD_def_or_no['Defaulted'] = PLD_def_or_no['LoanStatus'].apply(lambda x: 'Yes' if x in ["Chargedoff", "Defaulted"] else 'No')

# Convert the "Defaulted" column to a categorical variable
PLD_def_or_no['Defaulted'] = pd.Categorical(PLD_def_or_no['Defaulted'])

# Display the resulting DataFrame
#print(PLD_def_or_no)
```

```python
# Group data by the "Defaulted" column and calculate counts
defaulted_counts = PLD_def_or_no['Defaulted'].value_counts(normalize=True).sort_index()

# Bar plot for LoanStatus distribution
plt.figure(figsize=(5, 3))
plt.bar(defaulted_counts.index, defaulted_counts.values, color=['#B32B2B', '#006600'])

# Add percentage labels on top of bars
for i, count in enumerate(defaulted_counts):
    plt.text(i, count, f'{count * 100:.2f}%', ha='center', va='bottom', fontsize=12, fontweight='bold', color='white')

plt.xlabel("Defaulted")
plt.ylabel("Percentage")
plt.title("Defaulted vs. Non-Defaulted Loans")

plt.show()
```

3 out of 10 loans defaulted in the selected dataset. This is a really high percentage of defaulted loans. Lets plot the selected examined variables in the univariate plots section against the “Yes” and “No” categories.

```python
# Extract the year from the ListingCreationDate and create a new column ListingCreationYear
PLD_def_or_no['ListingCreationYear'] = pd.to_datetime(PLD_def_or_no['ListingCreationDate']).dt.year

# Calculate percentages of Defaulted vs. Non-Defaulted loans by year
percentage_df = (PLD_def_or_no.groupby(['ListingCreationYear', 'Defaulted']).size() / PLD_def_or_no.groupby(['ListingCreationYear']).size()).unstack()

# Reorder the columns to have "Yes" before "No"
percentage_df = percentage_df[['Yes', 'No']]

# Bar plot for Defaulted vs. Non-Defaulted loans by year with reversed order
plt.figure(figsize=(10, 6))
percentage_df.plot(kind='bar', stacked=True, color=['#B32B2B', '#006600'])

plt.xlabel("ListingCreationYear")
plt.ylabel("Percentage")
plt.title("Defaulted vs. Non-Defaulted Loans Over Time")
plt.legend(title='Defaulted', loc='upper left', labels=['Yes', 'No'], bbox_to_anchor=(1, 1))
plt.ylim([0, 1])  # Set the y-axis limit to show percentages between 0 and 1

plt.show()
```

```python
# Display a table of counts by year
table_by_year = PLD_def_or_no['ListingCreationYear'].value_counts().sort_index()
print("Counts by ListingCreationYear:")
print(table_by_year)
```

It seems that loans that were initiated in 2006 and 2007 had a higher chance of default than loans in other years. As a note keep in mind that for 2005 and 2014 there are few loan listings in the dataset (23 and 62 respectively).


```python
# Bar plot for Defaulted vs. Non-Defaulted loans by Term
plt.figure(figsize=(8, 6))
ax = PLD_def_or_no.groupby(['Term', 'Defaulted']).size().unstack().plot(kind='bar', stacked=True,
                                                                          color=['#006600', '#B32B2B'], width=0.6)
ax.set_xlabel("Term")
ax.set_ylabel("Proportion")
ax.set_title("Defaulted vs. Non-Defaulted Loans by Term")
ax.legend(title='Defaulted', labels=['No', 'Yes'])
ax.set_xticklabels(PLD_def_or_no['Term'].unique(), rotation=0)

# Calculate a contingency table
contingency_table = pd.crosstab(PLD_def_or_no['Defaulted'], PLD_def_or_no['Term'], margins=False)
result = stats.chi2_contingency(contingency_table)

print("Contingency Table:\n", contingency_table)
print("\nChi-Square Test Statistics:", result[0])
print("P-value:", result[1])
```

```python
# Calculate proportions of Defaulted vs. Non-Defaulted loans by Term
term_counts = PLD_def_or_no.groupby(['Term', 'Defaulted']).size().unstack()
term_proportions = term_counts.div(term_counts.sum(axis=1), axis=0)

# Bar plot for Defaulted vs. Non-Defaulted loans by Term with proportions on the y-axis
plt.figure(figsize=(8, 6))
ax = term_proportions.plot(kind='bar', stacked=True, color=['#006600', '#B32B2B'], width=0.6)
ax.set_xlabel("Term")
ax.set_ylabel("Percentage")
ax.set_title("Defaulted vs. Non-Defaulted Loans by Term")
ax.legend(title='Defaulted', labels=['No', 'Yes'])
ax.set_xticklabels(PLD_def_or_no['Term'].unique(), rotation=0)

# Calculate a contingency table
contingency_table = pd.crosstab(PLD_def_or_no['Defaulted'], PLD_def_or_no['Term'], margins=False)
result = stats.chi2_contingency(contingency_table)

print("Contingency Table:\n", contingency_table)
print("\nChi-Square Test Statistics:", result[0])
print("P-value:", result[1])

plt.show()
```

```python
# Calculate proportions of Defaulted vs. Non-Defaulted loans by Term
term_counts = PLD_def_or_no.groupby(['Term', 'Defaulted']).size().unstack()
term_proportions = term_counts.div(term_counts.sum(axis=1), axis=0)

# Bar plot for Defaulted vs. Non-Defaulted loans by Term with reversed order
plt.figure(figsize=(8, 6))
ax = term_proportions.loc[:, ['Yes', 'No']].plot(kind='bar', stacked=True,
                                                 color=['#B32B2B', '#006600'], width=0.6)
ax.set_xlabel("Term")
ax.set_ylabel("Percentage")
ax.set_title("Defaulted vs. Non-Defaulted Loans by Term")
ax.legend(title='Defaulted', labels=['Yes', 'No'], loc='upper right', bbox_to_anchor=(1.2, 1))
ax.set_xticklabels(PLD_def_or_no['Term'].unique(), rotation=0)
plt.show()
```



We can see that 36 months listings have a significant lower default rate than 12 and 60 months listings.

Let’s see if there is a relation between the loan amount and the default rate.

```python
# Create a violin plot for LoanOriginalAmount vs. Defaulted
plt.figure(figsize=(8, 6))
ax = sns.violinplot(data=PLD_def_or_no, x='Defaulted', y='LoanOriginalAmount', hue='Defaulted', palette=["#006600", "#B32B2B"])
ax.set_xlabel("Defaulted")
ax.set_ylabel("LoanOriginalAmount")
ax.set_title("Loan Amount vs. Defaulted")
ax.set_yticks(range(0, 35001, 5000))
plt.legend(title='Defaulted', labels=['No', 'Yes'])
```

```python

# Calculate summary statistics
summary_stats = PLD_def_or_no.groupby('Defaulted')['LoanOriginalAmount'].agg(
    median='median', mean='mean', standard_dev='std', IQR=lambda x: x.quantile(0.75) - x.quantile(0.25)).reset_index()

print("Summary Statistics:")
print(summary_stats)

plt.show()
```

It’s not likely that loan amount plays any significant role in the likelihood of default according to the above graph and summary statistics. Both violin plots looks pretty similar and their distribution is mainly concentrated below 5000 USD. The only difference that we can easily spot is that there are no loans above 25000 USD that had defaulted.
